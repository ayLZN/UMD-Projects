```python
#Importing the required libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import re
import seaborn as sns
from sklearn import linear_model
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
#Lists used to record song length and to test if lyrics were cleaned correctly
song_length = []
test = []
```

```python
#Reading csv file
df = pd.read_csv("dataset_8_lyrics.csv")
```

```python
for i in df['lyrics']:
    res = re.sub(r'[^\w\s]', '', i.lower()) #Cleaning by removing punctuation and lower casing strings
    test.append(res.split()) #Split every word to test
    song_length.append(len(res.split())) #Split every word and count the length of the song
```

```python
df['Song Length'] = song_length #Adding to the data frame
```

```python
sns.relplot(data=df, x="year", y='Song Length', kind='line') #Relplot
```

```python
#Converting values to 2d array
x1=df['year'].values.reshape(-1,1)
y1=df['Song Length'].values.reshape(-1,1)
```

```python
#Min Max Scaling
scaler1 = MinMaxScaler()
y_testt1 = scaler1.fit_transform(y1)
```

```python
xtrain1, xtest1, ytrain1, ytest1 = train_test_split(x1,y_testt1,test_size=.25)

model1=linear_model.LinearRegression() #Linear Regression
model1.fit(xtrain1,ytrain1)
```

```python
sns.regplot(x=xtrain1, y=ytrain1,line_kws={"color":"red"}) #To see correlation
```

```python
model1.coef_ #Printing correlation
```

```python
#list used to record the length of unique words
unique_length = []
```

```python
for x in df['lyrics']:
    res = re.sub(r'[^\w\s]', '', x.lower()) #Cleaning by removing punctuation and lower casing words
    i = np.array(res.split()) #Splitting words into array
    unique_length.append(len(np.unique(i))) #Counting length of unique words in array
```

```python
df['Unique Length'] = unique_length #Adding to the dataframe
```

```python
sns.relplot(data=df, x="year", y='Unique Length', kind='line') #Relplot
```

```python
#Converting into 2d array
x2=df['year'].values.reshape(-1,1)
y2=df['Unique Length'].values.reshape(-1,1)
```

```python
#Min Max Scaling
scaler2 = MinMaxScaler()
y_testt2 = scaler2.fit_transform(y2)
```

```python
xtrain2, xtest2, ytrain2, ytest2 = train_test_split(x2,y_testt2,test_size=.25)

model2=linear_model.LinearRegression() #Linear Regression
model2.fit(xtrain2,ytrain2)
```

```python
sns.regplot(x=xtrain2, y=ytrain2,line_kws={"color":"red"}) #To see correlation
```

```python
model2.coef_ #Printing correlation
```

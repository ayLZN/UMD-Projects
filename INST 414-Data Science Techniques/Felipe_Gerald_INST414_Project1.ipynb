```python
#Creating a list for all the urls
urls = []
#For loop to append each lyric url to the new list
for lyric_url in dataset_8["lyricsurl"]:
    urls.append(lyric_url)
```

```python
#testing with 2 urls
url_test = ["https://www.azlyrics.com/lyrics/gangoffour/outsidethetrainsdontrunontime.html","https://www.azlyrics.com/lyrics/engelberthumperdinck/allthiswordandthesevenseas.html"]
```

```python
#This was used to test the 2 urls (one gave me 404 error)

#Creating a loop to iterate through the list of urls
lyrics_test_list = [] #New list to hold actual lyrics
counter_test = 0 #Used to see the # of lyrics
for lyric_url in url_test:
    #Getting the data from the urls list
    counter_test += 1
    pages_test = requests.get(lyric_url) #Getting the page of the lyric url
    print(counter_test, pages_test.status_code, lyric_url) #Prints out #, the status code and the url after acquiring using requests.get
    beautiful_soup = BeautifulSoup(pages_test.content, "lxml") #Reading the content from the page
    info = beautiful_soup.find(class_="col-xs-12 col-lg-8 text-center") #Getting the content from the page

    #Extracting the info after using beautifulsoup
    information_list_test = info.text.split("\n")
    
    if "\r" in information_list_test:
        #deleteing anything before \r
        r_index = information_list_test.index("\r")
        del information_list_test[:(r_index + 1)]
    if " Submit Corrections" in information_list_test:
        #deleteing anything after submit corrections for a cleaner list
        submit_index = information_list_test.index(" Submit Corrections")
        del information_list_test[submit_index:]

    #Taking out any blank items that made it to the list
    lyrics_test = [lyrics_test for lyrics_test in information_list_test if lyrics_test]

    #Appending to the lyric list
    lyrics_test_list.append(lyrics_test)
    #60 second pauses to not get blocked from the website
    time.sleep(60)
```

```python
#Creating a loop to iterate through the list of urls
lyrics_list = [] #New list to hold actual lyrics
counter = 0 #Used to see the # of lyrics
for lyric_url in urls:
    #Getting the data from the urls list
    counter += 1
    pages = requests.get(lyric_url) #Getting the page of the lyric url
    print(counter, pages.status_code, lyric_url) #Prints out #, the status code and the url after acquiring using requests.get
    beautiful_soup = BeautifulSoup(pages.content, "lxml") #Reading the content from the page
    info = beautiful_soup.find(class_="col-xs-12 col-lg-8 text-center") #Getting the content from the page

    #Extracting the info after using beautifulsoup
    information_list = info.text.split("\n")
    
    if "\r" in information_list:
        #deleteing anything before \r
        r_index = information_list.index("\r")
        del information_list[:(r_index + 1)]
    if " Submit Corrections" in information_list:
        #deleteing anything after submit corrections for a cleaner list
        submit_index = information_list.index(" Submit Corrections")
        del information_list[submit_index:]

    #Taking out any blank items that made it to the list
    lyrics = [lyrics for lyrics in information_list if lyrics]

    #Appending to the lyric list
    lyrics_list.append(lyrics)
    #60 second pauses to not get blocked from the website
    time.sleep(60)
```

```python
#confirming all 499 songs were appended
print(len(lyrics_list))
```

```python
#To see the lyrics collection list
lyrics_list
```

```python
#Zipping lyrics
lyric_table = list(zip(*(dataset_8["songtitle"],dataset_8["artist/bandname"], dataset_8["year"], urls, lyrics_list)))
```

```python
#Turning the list into a csv
#Opening the csv and writing in it
with open('dataset_8_lyrics.csv','w',encoding = "utf-8", newline = '') as collected_lyrics:
    wr = csv.writer(collected_lyrics)
    wr.writerow(("songtitle","artist/bandname", "year", "lyricsurls", "lyrics"))
    wr.writerows(lyric_table)
collected_lyrics.close()
```
